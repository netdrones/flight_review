```python
estimator_local_position = ulog.get_dataset('estimator_local_position').data
eph = estimator_local_position['eph']
epv = estimator_local_position['epv']
plt.plot(eph, color='blue', label='eph')
plt.plot(epv, color='green', label='epv')
plt.title('Local Position Estimate (EKF)')
plt.xlabel('Position x (m)')
plt.ylabel('Position y (m)')
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
